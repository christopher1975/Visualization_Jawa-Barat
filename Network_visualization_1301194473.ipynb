```python
# import library

import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Membuat Graph

G = nx.Graph()
G.add_edges_from([("Bogor", "Bekasi"), ("Bogor", "Sukabumi"), ("Bogor", "Cianjur")])
G.add_edges_from([("Sukabumi", "Cianjur")])
G.add_edges_from([("Cianjur", "Garut"), ("Cianjur", "Bandung")])
G.add_edges_from([("Bekasi", "Karawang")])
G.add_edges_from([("Karawang", "Purwakarta"), ("Karawang", "Subang")])
G.add_edges_from([("Purwakarta", "Bandung"), ("Purwakarta", "Subang")])
G.add_edges_from([("Bandung", "Garut"), ("Bandung", "Sumedang"), ("Bandung", "Subang")])
G.add_edges_from([("Subang", "Sumedang"), ("Subang", "Indramayu")])
G.add_edges_from([("Sumedang", "Garut"), ("Sumedang", "Majalengka"), ("Sumedang", "Indramayu")])
G.add_edges_from([("Garut", "Tasikmalaya")])
G.add_edges_from([("Tasikmalaya", "Ciamis"), ("Tasikmalaya", "Banjar")])
G.add_edges_from([("Majalengka", "Indramayu"), ("Majalengka", "Cirebon"), ("Majalengka", "Kuningan"), ("Majalengka", "Ciamis")])
G.add_edges_from([("Indramayu", "Cirebon")])
G.add_edges_from([("Ciamis", "Kuningan"), ("Ciamis", "Banjar")])
G.add_edges_from([("Kuningan", "Cirebon")])

plt.figure(figsize=(20, 15))
plt.title("Jalur Darat Antar Kota Pada Provinsi Jawa Barat", size=20)
nx.draw(G, with_labels = True, node_size=5000, node_color='c')
plt.show()
```

```python
# Membuat Degree Centrality

Degree = nx.degree_centrality(G)
dcs = pd.DataFrame(sorted(Degree.items(),key=lambda x: x[1], reverse = True))
dcs.columns = ['Kota', 'Degree Centrality']
dcs
```

```python
# Membuat Betweenness Centrality

Betweenness = nx.betweenness_centrality(G)
bcs = pd.DataFrame(sorted(Betweenness.items(),key=lambda x: x[1], reverse = True))
bcs.columns = ['Kota', 'Betweenness Centrality']
bcs
```

```python
# Menentukan Kota Terpenting Berdasarkan Nilai Degree Centrality dan Betweenness Centrality

hasil = dcs.merge(bcs, on="Kota")
hasil["Hasil Rata Rata"] = (hasil["Degree Centrality"] + hasil["Betweenness Centrality"]) / 2
del hasil["Degree Centrality"]
del hasil["Betweenness Centrality"]
hasil = hasil.sort_values(by="Hasil Rata Rata", ascending=False)
hasil.style.hide_index()
```

```python
#Visualisasi Kota Terpenting

color = []
for node in G:
  if node == hasil.reset_index()["Kota"][0]:
    color.append('green')
  else:
    color.append('c')
plt.figure(figsize=(20, 15))
plt.title("Jalur Darat Antar Kota Pada Provinsi Jawa Barat", size=20)
nx.draw(G, with_labels = True, node_size=5000, node_color=color)
plt.show()
```

Visualisasi data menunjukan bahwa kota terpenting dari jalur darat antar kota pada provinsi Jawa Barat adalah kota Sumedang dengan warna node hijau
